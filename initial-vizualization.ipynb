```python
import pandas as pd
df = pd.read_csv("Cleaned_Quality_of_Service_5G.csv")
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.countplot(data=df, y="Application_Type", order=df["Application_Type"].value_counts().index)
plt.title("Distribution of Application Types")
plt.xlabel("Count")
plt.ylabel("Application Type")
plt.show()
```

```python
sns.histplot(df["Signal_Strength_dBm"], bins=20, kde=True)
plt.title("Distribution of Signal Strength (dBm)")
plt.xlabel("Signal Strength (dBm)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(data=df, x="Required_Bandwidth_Mbps", y="Latency_ms", hue="Application_Type")
plt.title("Bandwidth vs Latency by Application Type")
plt.xlabel("Required Bandwidth (Mbps)")
plt.ylabel("Latency (ms)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
numeric_df = df.select_dtypes(include=["float64", "int64"])
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
